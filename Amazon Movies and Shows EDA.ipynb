```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('amazon_prime_titles.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['director'].replace({np.nan : 'unavailable'},inplace=True)
```

```python
df['cast'].fillna('unavailable',inplace=True)
```

```python
df['country'].value_counts()
```

```python
df['country'].fillna(df['country'].mode()[0],inplace=True)
```

```python
df['date_added'].ffill(inplace=True)
```

```python
df.info()
```

```python
df['rating'].fillna(df['rating'].mode()[0],inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],format='%B %d, %Y')
```

```python
df['date_added']
```

```python
df['rating'].unique()
```

```python
df['rating'].replace(
    {'AGES_16_' : '16+',
     '16' : '16+',
     'AGES_18_' : '18+',
     'NOT_RATE' : 'UNRATED',
     'ALL_AGES' : 'ALL'
     } , inplace=True
)
```

```python
df['rating'].unique()
```

```python
df['type'].unique()
```

```python
df['type'].value_counts()
```

```python
df.rename({'listed_in':'categories'},inplace=True,axis=1)
```

```python
df.head()
```

```python
df['duration'].unique()
```

```python
df['duration_min']=df['duration'].apply(lambda x:int(x.split()[0]) if 'min' in x else 0)
```

```python
df['duration_season']=df['duration'].apply(lambda x:int(x.split()[0]) if ('Season' or 'Seasons') in x else 0)
```

```python
df.drop('duration',inplace=True,axis=1)
```

```python
df.head()
```

```python
df['month']=df['date_added'].dt.month
```

```python
df['year']=df['date_added'].dt.year
```

```python
df.head()
```

```python
pd.crosstab(df['country'],df['type'])
```

```python
df.groupby('categories')['duration_min'].mean().sort_values(ascending=False)
```

```python
df.to_csv('Amaaazon.csv',index=False,encoding='utf-8')
```
